## Annual Result

```python
# write the query
query = """
SELECT
	order_year AS year,
	SUM(var_rev_seller) AS var_rev,
	SUM(fix_rev_seller) AS fix_rev,
	SUM(rep_cost) AS reput_cost,
	SUM(order_count) AS n_order
FROM dre_table_year
GROUP BY order_year;
"""
# Execute the query
c.execute(query)
#Fetch and print the results
c.fetchall()
df = pd.read_sql_query(query, conn)
```

```python
df['it_cost'] = round(((47.8/(df['n_order']**(1/2)))+2.2)*df['n_order'],2)
df['total_rev'] = round(df['var_rev']+df['fix_rev'],2)
df['total_cost'] = round(df['it_cost']+df['reput_cost'],2)

df['result'] = round(df['total_rev']-df['total_cost'],2)
```

## Sellers Revenue

```python
# write the query
query = """
SELECT
	seller_id,
	SUM(var_rev_seller + fix_rev_seller - rep_cost) AS res_seller
FROM dre_table_year
WHERE order_year != "2016"
GROUP BY seller_id
ORDER BY SUM(var_rev_seller + fix_rev_seller - rep_cost) DESC 
"""
# Execute the query
c.execute(query)
#Fetch and print the results
c.fetchall()
df4 = pd.read_sql_query(query, conn)
```

```python
# Define bin edges with a constant width of 200, then adjust the final bin to include 2500 or more
bin_edges = np.arange(df4['res_seller'].min(), df4['res_seller'].max(), 500)
# bin_edges = np.append(constant_bin_edges, df4['res_seller'].max())

# Plot histogram with specific bin ranges
plt.figure(figsize=(10, 6))
plt.hist(df4['res_seller'], bins=bin_edges, edgecolor='k')
plt.title('Histogram of res_seller with Custom Bin Ranges')
plt.xlabel('res_seller')
plt.ylabel('Frequency')
plt.show()
```

## Geolocation 

```python
# write the query
query = """
SELECT
	  seller_state,
	  customer_state,
	  COUNT(order_id)
  FROM geo_loc_det
  WHERE order_year != "2016"
  GROUP BY seller_state, customer_state
  ORDER BY COUNT(order_id) DESC , seller_state, customer_state
"""
# Execute the query
c.execute(query)
#Fetch and print the results
c.fetchall()
df5 = pd.read_sql_query(query, conn)
```

```python
# Aggregate COUNT by seller_state
seller_state_counts = df5.groupby('seller_state')['COUNT(order_id)'].sum().reset_index()

# Define a blue color palette
colors = plt.cm.Blues([i/float(len(seller_state_counts['COUNT(order_id)'])-1) for i in range(len(seller_state_counts['COUNT(order_id)']))])

# Plotting the treemap
plt.figure(figsize=(12, 8))
squarify.plot(sizes=seller_state_counts['COUNT(order_id)'], label=seller_state_counts['seller_state'], color=colors, alpha=0.8)
plt.title("Treemap of COUNT per Seller State (Blue Palette)")
plt.axis('off')
plt.show()
```

```python
# write the query
query = """
SELECT
	  gld.seller_state,
	  gld.customer_state,
	  ROUND(SUM(sid.price)*0.1,0) AS var_rev,
	  COUNT(gld.order_id) AS quant_ord
FROM geo_loc_det gld
INNER JOIN sales_item_detail sid ON sid.order_id = gld.order_id 
WHERE gld.order_year != "2016"
GROUP BY gld.seller_state, gld.customer_state
ORDER BY SUM(sid.price)*0.1 DESC
"""
# Execute the query
c.execute(query)
#Fetch and print the results
c.fetchall()
df6 = pd.read_sql_query(query, conn)
```

```python
# Aggregating data by seller_state
aggregated_data = df6.groupby('seller_state').sum()

# Sorting the aggregated data by quant_ord in descending order
aggregated_data_sorted = aggregated_data.sort_values(by='quant_ord', ascending=False)

# Re-plotting with sorted data and dual y-axes

fig, ax1 = plt.subplots(figsize=(10, 6))

# Setting up the positions and width for grouped bars
x = np.arange(len(aggregated_data_sorted.index))  # label locations
width = 0.35  # width of the bars

# Bar plot for var_rev on the primary y-axis
bar1 = ax1.bar(x - width/2, aggregated_data_sorted['var_rev'], width, label='Revenue (var_rev)', color='b', alpha=0.7)
ax1.set_xlabel('Seller State')
ax1.set_ylabel('Revenue (var_rev)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Creating a second y-axis for quant_ord
ax2 = ax1.twinx()
bar2 = ax2.bar(x + width/2, aggregated_data_sorted['quant_ord'], width, label='Quantity Ordered (quant_ord)', color='g', alpha=0.7)
ax2.set_ylabel('Quantity Ordered (quant_ord)', color='g')
ax2.tick_params(axis='y', labelcolor='g')

# Adding title and setting x-ticks
ax1.set_xticks(x)
ax1.set_xticklabels(aggregated_data_sorted.index)
plt.title('Grouped Bar Plot of Revenue and Quantity Ordered per Seller State (Ordered by Quantity)')

# Adjusting layout for clarity
fig.tight_layout()
plt.show()
```

# Customer Satisfaction ðŸ˜

```python
# write the query
query = """
SELECT
	sid.product_category_name,
	ROUND(AVG(sid.customer_review),1) AS aver_score,
	ROUND(SUM(dtyc.var_rev_seller + dtyc.fix_rev_seller - dtyc.rep_cost),0) AS rev_sell
FROM sales_item_detail sid
INNER JOIN dre_table_year_cat dtyc ON dtyc.seller_id = sid.seller_id 
WHERE sid.order_year != "2016"
AND sid.product_category_name IS NOT NULL
GROUP BY sid.product_category_name
ORDER BY AVG(sid.customer_review) DESC
"""
# Execute the query
c.execute(query)
#Fetch and print the results
c.fetchall()
df7 = pd.read_sql_query(query, conn)
```

```python
plt.figure(figsize=(12, 8))
scatter_plot = plt.scatter(
    df7['aver_score'], df7['product_category_name'],
    c=df7['rev_sell'], s=df7['rev_sell'] / 10000,  # Scale size by revenue for visibility
    cmap='viridis', alpha=0.6, edgecolors="w", linewidth=0.5
)

plt.colorbar(scatter_plot, label='Revenue (rev_sell)')
plt.title('Scatter Plot of Product Categories by Average Score and Revenue')
plt.xlabel('Average Score')
plt.ylabel('Product Category Name')
plt.show()
```

```python
# write the query
query = """
SELECT
	sid.seller_state,
	ROUND(AVG(sid.customer_review),1) AS aver_score,
	ROUND(SUM(dtyc.var_rev_seller + dtyc.fix_rev_seller - dtyc.rep_cost),0) AS rev_sell
FROM sales_item_detail sid
INNER JOIN dre_table_year_cat dtyc ON dtyc.seller_id = sid.seller_id 
WHERE sid.order_year != "2016"
AND sid.product_category_name IS NOT NULL
GROUP BY sid.seller_state
ORDER BY AVG(sid.customer_review) DESC
"""
# Execute the query
c.execute(query)
#Fetch and print the results
c.fetchall()
df8 = pd.read_sql_query(query, conn)
```

```python
plt.figure(figsize=(12, 8))
scatter_plot = plt.scatter(
    df8['aver_score'], df8['seller_state'],
    c=df8['rev_sell'], s=df8['rev_sell'] / 10000,  # Scale size by revenue for visibility
    cmap='viridis', alpha=0.6, edgecolors="w", linewidth=0.5
)

plt.colorbar(scatter_plot, label='Revenue (rev_sell)')
plt.title('Scatter Plot of Product Categories by Average Score and Revenue')
plt.xlabel('Average Score')
plt.ylabel('Product Category Name')
plt.show()
```
